```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
amp = 1; sr = 10000; phase = 0
E  = 659.25
D  = 587.33
C  = 523.25
G = 783.99

s_all = np.empty(0)
```

```python
notes=[E,D,C,D,E,E,E,D,D,D,E,G,G, \
      E,D,C,D,E,E,E,D,D,E,D,C]
durs=[0.45,0.25,0.25,0.25,0.25,0.25,0.3,0.25,0.25,0.3,0.25,0.25,0.3, \
      0.45,0.25,0.25,0.25,0.25,0.25,0.3,0.25,0.25,0.25,0.25,0.3]
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```

```python
ipd.Audio(s_all, rate=sr)
```
